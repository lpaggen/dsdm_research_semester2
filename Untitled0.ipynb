```python
import re
import pandas as pd
import os
import glob
import json
from collections import defaultdict
from google.colab import drive

# ──────────────────── Step 1: TXT-Based Keyword Counting ────────────────────

categories = {
    'quantum computing': ['quantum computing'],
    'error_correction': ['error correction'],
    'surface_code': ['surface code'],
    'error_mitigation': ['error mitigation'],
    'fault_tolerant': [
        'fault-tolerant algorithm', 'fault tolerant algorithm',
        'fault tolerant', 'fault-tolerant', 'fault-tolerant quantum'
    ],
    'ai_quantum': ['ai', 'artificial intelligence', 'quantum']  # AI + Quantum both must appear
}

folder_path = "/content/drive/MyDrive/Colab Notebooks/Data"
year_keyword_counts = {}
txt_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.txt')]

for filename in txt_files:
    match = re.search(r'_(20\d{2})_', filename)
    if not match:
        continue
    year = int(match.group(1))

    with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as f:
        text = f.read().lower()

    if year not in year_keyword_counts:
        year_keyword_counts[year] = {cat: 0 for cat in categories}

    for cat, keywords in categories.items():
        if cat == 'ai_quantum':
            if any(k in text for k in ['ai', 'artificial intelligence']) and 'quantum' in text:
                year_keyword_counts[year][cat] += 1
        else:
            if any(k in text for k in keywords):
                year_keyword_counts[year][cat] += 1

df_txt = pd.DataFrame.from_dict(year_keyword_counts, orient='index').sort_index()
df_txt.index.name = "Year"

# ──────────────────── Step 2: JSON-Based Metadata Loading ────────────────────

drive.mount('/content/drive')
json_dir = '/content/drive/My Drive/Jsons'
json_files = glob.glob(f'{json_dir}/*.json')

records = []
for fn in json_files:
    m = re.search(r'(19|20)\d{2}', os.path.basename(fn))
    year = int(m.group(0)) if m else None

    with open(fn, 'r', encoding='utf-8') as f:
        data = json.load(f)

    for title, info in data.items():
        keywords = info.get('keywords', [])
        locations = info.get('locations', [])
        if year is None or not keywords or not locations:
            continue
        records.append({
            'title': title,
            'year': year,
            'keywords': ';'.join(keywords),
            'locations': ';'.join(locations)
        })

df_json_raw = pd.DataFrame(records)

# ──────────────────── Step 3: JSON-Based Document-Level Keyword Counting ────────────────────

category_keywords = {
    'quantum computing': ['quantum computing'],
    'error_correction': ['error correction'],
    'surface_code': ['surface code'],
    'error_mitigation': ['error mitigation'],
    'fault_tolerant': [
        'fault-tolerant algorithm', 'fault tolerant algorithm',
        'fault tolerant', 'fault-tolerant', 'fault-tolerant quantum'
    ],
    'ai_quantum': ['ai', 'artificial intelligence']  # quantum will be handled separately
}

json_counts = defaultdict(lambda: defaultdict(int))

# Group keywords per document (year, title)
doc_keywords = (
    df_json_raw
    .assign(keywords=df_json_raw['keywords'].str.lower().str.split(';'))
    .explode('keywords')
    .dropna(subset=['keywords'])
    .assign(keywords=lambda d: d['keywords'].str.strip())
    .groupby(['year', 'title'])['keywords']
    .apply(list)
    .reset_index()
)

for _, row in doc_keywords.iterrows():
    year = row['year']
    keywords = row['keywords']

    for cat, kw_list in category_keywords.items():
        if cat != 'ai_quantum':
            if any(any(kw in k for kw in kw_list) for k in keywords):
                json_counts[cat][year] += 1

    # Special case for ai_quantum: must include both ai and quantum
    has_ai = any(k in ['ai', 'artificial intelligence'] for k in keywords)
    has_quantum = any('quantum' in k for k in keywords)
    if has_ai and has_quantum:
        json_counts['ai_quantum'][year] += 1

df_json = pd.DataFrame(json_counts).fillna(0).astype(int)
df_json.index.name = "Year"

# ──────────────────── Step 4: Combine TXT + JSON Counts ────────────────────

# Reindex both DataFrames to ensure same year range
all_years = pd.RangeIndex(start=2010, stop=2025)
df_txt = df_txt.reindex(all_years, fill_value=0)
df_json = df_json.reindex(all_years, fill_value=0)

# Combine
df_combined = df_txt.copy()
for col in df_json.columns:
    df_combined[f"{col}_json"] = df_json[col]

df_combined = df_combined.loc[2014:2024]

# Display result
df_combined.head(10)
```

```python
categories = ['error_correction', 'surface_code', 'error_mitigation', 'fault_tolerant', 'ai_quantum', 'quantum computing']

# Dynamically merge only if both base and _json columns exist
df_merged = df_combined.copy()

for cat in categories:
    json_col = f"{cat}_json"
    if cat in df_merged.columns and json_col in df_merged.columns:
        df_merged[cat] = df_merged[cat] + df_merged[json_col]
    elif json_col in df_merged.columns:
        df_merged[cat] = df_merged[json_col]  # only json exists
    # else: keep existing .txt-only column or skip

# Drop _json columns if they exist
cols_to_drop = [f"{cat}_json" for cat in categories if f"{cat}_json" in df_merged.columns]
df_merged.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

X_years = df_merged.index.to_numpy().reshape(-1, 1)
pred_year = np.array([[2025]])

predictions_rf = {}

print("🌲 Random Forest Regressor")
for column in df_merged.columns:
    y_values = df_merged[column].values

    # Split: Train (2014–2022), Validate (2023–2024)
    X_train, y_train = X_years[:9], y_values[:9]
    X_val, y_val = X_years[9:], y_values[9:]

    # Fit random forest
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Predict 2025
    y_pred_2025 = model.predict(pred_year)[0]
    predictions_rf[column] = max(0, round(y_pred_2025))

    # Validation RMSE
    val_pred = model.predict(X_val)
    val_rmse = np.sqrt(mean_squared_error(y_val, val_pred))
    print(f"{column.title():<18} → Pred 2025: {y_pred_2025:.1f} | Val RMSE: {val_rmse:.2f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Prepare years
X_years = df_merged.index.to_numpy().reshape(-1, 1)
future_years = np.array([[2025], [2026]])

predictions_poly = {}

print("🔁 Polynomial Regression (Degree 2) – Predict 2025 & 2026")
for column in df_merged.columns:
    y_values = df_merged[column].values

    X_train, y_train = X_years[:9], y_values[:9]
    X_val, y_val = X_years[9:], y_values[9:]

    model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
    model.fit(X_train, y_train)

    # Predict both future years
    future_preds = model.predict(future_years)
    predictions_poly[column] = [max(0, round(p)) for p in future_preds]

    val_pred = model.predict(X_val)
    val_rmse = np.sqrt(mean_squared_error(y_val, val_pred))
    print(f"{column.title():<18} → Pred 2025: {future_preds[0]:.1f}, 2026: {future_preds[1]:.1f} | Val RMSE: {val_rmse:.2f}")

```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Normalize years to start from 0 (i.e., 2014 = 0)
X_years = (df_merged.index.to_numpy() - 2014).reshape(-1, 1)
future_years = np.array([[2025 - 2014], [2026 - 2014]])

predictions_poly2 = {}

print("🔁 Improved Polynomial Regression (Degree 2 + Ridge)")
for column in df_merged.columns:
    y_values = df_merged[column].values
    X_train, y_train = X_years[:9], y_values[:9]
    X_val, y_val = X_years[9:], y_values[9:]

    model = make_pipeline(PolynomialFeatures(degree=2), Ridge(alpha=2))
    model.fit(X_train, y_train)

    # Predict 2025 and 2026
    y_pred_future = model.predict(future_years)
    predictions_poly2[column] = [max(0, round(p)) for p in y_pred_future]

    val_pred = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, val_pred))

    print(f"{column.title():<18} → Pred 2025: {y_pred_future[0]:.1f}, 2026: {y_pred_future[1]:.1f} | Val RMSE: {rmse:.2f}")
```
